### What are the top 10 monitored products by State?


**1.- Get the list of top monitored products**

```python
HMonitored =df['producto'].value_counts().sort_values(ascending=False).reset_index()
HMonitored.columns = ['producto', 'conteo']
HMonitored = HMonitored[:10]
HMonitored
```

**2.- Filter dataframe in order to get only the rows related to the High Monitored Products**

```python
Filtered_HM = df[df.producto.isin(HMonitored['producto'])]
```

**3.- Get High Monitored products by state**

```python
dfl = pd.crosstab(index=Filtered_HM['producto'], columns=Filtered_HM['estado']).sort_values('producto', ascending=True)
dfl
```

## Use the data to find an interesting fact


**Commercial chains do not influence in the product monitored number. this is accepted because there is not estatistical significance**

```python
test_table_1 = pd.crosstab(index=Filtered_HM['cadenaComercial'], columns=Filtered_HM['producto'], margins=True)
```

```python
from scipy.stats import chi2_contingency
test1 = chi2_contingency(test_table_1)
print(test1[1])
```

**The category have influence in the products monitored number, we can see there is an statistical significance and this happen because  there are different products with same category.**

```python
test_table_2 = pd.crosstab(index=Filtered_HM['categoria'], columns=Filtered_HM['producto'], margins=True)
test_table_2
```

```python
from scipy.stats import chi2_contingency
test2 = chi2_contingency(test_table_2)
print(test2[1])
```
